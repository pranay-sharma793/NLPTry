```python
import matplotlib.pyplot as plt
import numpy as np

# Data
combinations = [
    'Swedish + Norwegian',
    'Swedish + Norwegian + Danish',
    'Swedish + Norwegian + Spanish + French',
    'Swedish + Norwegian + Spanish + French + Arabic + Hindi',
    'Swedish + Norwegian + Spanish + French + Arabic + Hindi + Danish'
]

f1_values = [0.393873085, 0.359133127, 0.367630058, 0.335329341, 0.3477715]
precision_values = [0.481283422, 0.481327801, 0.489230769, 0.425855513, 0.427469136]
recall_values = [0.333333333, 0.286419753, 0.294444444, 0.27654321, 0.293121693]
x_labels = ['C1', 'C2', 'C3', 'C4', 'C5']

bar_width = 0.2
index = np.arange(len(combinations))

# Colors
f1_color = '#6fa3d1'
precision_color = '#fae20c'
recall_color = '#fa912f' 

plt.figure(figsize=(6, 7))

# Plotting
bars1 = plt.bar(index, f1_values, width=bar_width, label='F1', color=f1_color, edgecolor='grey', linewidth=0.5)
bars2 = plt.bar(index + bar_width, precision_values, width=bar_width, label='Precision', color=precision_color, edgecolor='grey', linewidth=0.5)
bars3 = plt.bar(index + 2 * bar_width, recall_values, width=bar_width, label='Recall', color=recall_color, edgecolor='grey', linewidth=0.5)


# Customizing the plot
plt.xlabel('Combinations')
plt.ylabel('Scores')
# plt.title('Evaluation Metrics for Different Combinations')
plt.xticks(index + bar_width, x_labels)

plt.margins(y=0.2)

# Adding legends
legend1 = plt.legend(handles=[bars1[0], bars2[0], bars3[0]], labels=['F1', 'Precision', 'Recall'], loc='upper right', ncol=1)
plt.gca().add_artist(legend1)

# Generate combined labels for legend2
combined_labels = [f'{label}: {combination}' for label, combination in zip(x_labels, combinations)]

# Adding legend2 to the axes
legend2 = plt.legend(bars1, combined_labels, title='Combinations', loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=1)

# Show the plot
plt.tight_layout()
plt.show()
```
