Translation

```python
import pandas as pd
import string
from googletrans import Translator

translator = Translator()
def translate_first_element(element_list, target_language='hi'):
    if not element_list:
        return element_list

    first_element = element_list[0][0]
    translated_first_element=translator.translate(first_element, target_language).text
    element_list[0][0] = translated_first_element

    return element_list

def translate_tokens(tokens, target_language='hi'):
    translated_tokens = [translator.translate(token, dest=target_language).text for token in tokens]
    translated_tokens = [''.join(char for char in token if char not in string.punctuation) for token in translated_tokens]
    # print[translated_tokens]
    return translated_tokens

def translate_sentence(sentence, target_language='hi'):
    translated_sentence = translator.translate(sentence, dest=target_language).text
    return translated_sentence

def add_quotes(series):
    # Remove single quotes from each element in the series
    series = series.apply(lambda x: [item.replace("'", "") for item in x])

    # Add single quotes back to each element in the series
    series = series.apply(lambda x: [f"'{item}'" for item in x])

    return series

def translate(df, target_language='hi'):

    # Extract the 'tokens' column
    tokens_column = df['tokens']

    # Translate the tokens to a target language (default is English)
    translated_tokens = tokens_column.apply(translate_tokens)

    df['tokens'] = add_quotes(translated_tokens)

    # df['tokens'] = translated_tokens

    # Combine the translated tokens into a sentence
    translated_sentences = [' '.join(tokens) for tokens in translated_tokens]

    # Translate the 'ori_aspect' column
    df['ori_aspect'] = df['ori_aspect'].apply(translate_sentence)

    # Translate only the first element in the 'aspect_sentiment' column
    df['aspect_sentiment'] = df['aspect_sentiment'].apply(translate_first_element)

    # Create a new column with translated sentences
    df['sentence'] = translated_sentences
    df['ori_sentence'] = df['sentence']

    return df

file_path = '/content/drive/MyDrive/CS678Project/multilinguality/test_rospacy_sampled.json'
df = pd.read_json(file_path)

df  = translate(df)

translated_json_file_path = '/content/drive/MyDrive/CS678Project/multilinguality/hindi_translation.json'
df.to_json(translated_json_file_path, orient='records')




```
